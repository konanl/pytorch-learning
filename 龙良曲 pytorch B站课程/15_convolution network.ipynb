```python
# With square kernels and equal stride
m1 = torch.nn.Conv2d(16, 33, 3, stride=2)
# non-square kernels and unequal stride and with padding
m2 = torch.nn.Conv2d(16, 33, (3, 5), stride=(2, 1), padding=(4, 2))
# non-square kernels and unequal stride and with padding and dilation
m3 = torch.nn.Conv2d(16, 33, (3, 5), stride=(2, 1), padding=(4, 2), dilation=(3, 1))
input = torch.randn(20, 16, 50, 100)
output1 = m1(input)
output2 = m2(input)
output3 = m3(input)
```

```python
output1.shape, output2.shape, output3.shape
```

```python
# DownSample
# .Max Pooling
# .Average Pooling

# max pooling 
x = torch.Tensor(1,16,14,14)
print(x.shape)
layer = nn.MaxPool2d(2,stride=2)
out = layer(x)
print(out.shape)
```

```python
# average pooling
layer0 = nn.AvgPool2d(2,stride=2)
out = layer0(x)
print(out.shape)
```
